```python
!pip install language-tool-python
import language_tool_python

tool = language_tool_python.LanguageTool('en-US')
def grammar_score(sentence):
    matches = tool.check(sentence)
    score = 1 - (len(matches) / max(len(sentence.split()), 1))
    return max(score, 0) * 100  # Convert to percentage

# Create lists to store the grammar scores
question_scores = []
average_variation_scores = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    # Grammar score for the question
    question = row['question']
    question_score = grammar_score(question)
    question_scores.append(question_score)

    # Grammar scores for variations and calculate the average
    variation_scores = [grammar_score(row[f'variation_{i}']) for i in range(1, 11)]
    avg_variation_score = sum(variation_scores) / 10
    average_variation_scores.append(avg_variation_score)

    # Optional: Print results for each row
    print(f"Row {index + 1}:")
    print(f"Question Grammar Score: {question_score:.2f}%")
    print(f"Average Variation Grammar Score: {avg_variation_score:.2f}%\n")

# Add the scores to the DataFrame
df['question_grammar_score'] = question_scores
df['average_variation_grammar_score'] = average_variation_scores

# Print final DataFrame with scores
print(df)

# Optional: Save the DataFrame to a CSV file
df.to_csv('grammar_scores.csv', index=False)
```
